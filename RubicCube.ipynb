```python
import numpy as np
# define the state 

state_initial = [
['w','y','o','b','r','w','o','r','r'],
[
        ['b','o','y','w','y','r','o','o','g'],   # l
        ['g','g','g','b','b','g','w','y','o'],   # f
        ['y','b','w','w','w','o','y','y','r'],   # r
        ['b','r','r','w','g','r','b','y','g']    # b
],
['r','o','b','g','o','b','w','g','y']
]


# define movement (all clock wise facing the side)

def u(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [u[:3], u[3:6], u[6:]]
    m = np.array(side_m)
    u_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = l[:]
    l[:3] = f[:3]
    f[:3] = r[:3]
    r[:3] = b[:3]
    b[:3] = tmp[:3]
    
    result = [u_new, [l,f,r,b], d]
    return result

def d(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [d[:3], d[3:6], d[6:]]
    m = np.array(side_m)
    d_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = l[:]
    l[6:] = b[6:]
    b[6:] = r[6:]
    r[6:] = f[6:]
    f[6:] = tmp[6:]
    
    result = [u, [l,f,r,b], d_new]
    return result

def f(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [f[:3], f[3:6], f[6:]]
    m = np.array(side_m)
    f_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = l[:]
    l[2], l[5], l[8] = d[:3]
    d[:3] = r[0], r[3], r[6]
    r[0], r[3], r[6] = u[6:]
    u[6:] = tmp[8], tmp[5], tmp[2]
    
    result = [u, [l,f_new,r,b], d]
    return result
    
def b(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [b[:3], b[3:6], b[6:]]
    m = np.array(side_m)
    b_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = l[:]
    l[6], l[3], l[0] = u[:3]
    u[:3] = r[2], r[5], r[8]
    r[8], r[5], r[2] = d[6:]
    d[6:] = tmp[0], tmp[3], tmp[6]
    
    result = [u, [l,f,r,b_new], d]
    return result

def l(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [l[:3], l[3:6], l[6:]]
    m = np.array(side_m)
    l_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = f[:]
    f[0], f[3], f[6] = u[0], u[3], u[6]
    u[0], u[3], u[6] = b[8], b[5], b[2]
    b[8], b[5], b[2] = d[0], d[3], d[6]
    d[0], d[3], d[6] = tmp[0], tmp[3], tmp[6]
    
    result = [u, [l_new,f,r,b], d]
    return result

def r(state):
    l,f,r,b = state[1]
    u = state[0]
    d = state[2]
    
    side_m = [r[:3], r[3:6], r[6:]]
    m = np.array(side_m)
    r_new = [item for sublist in np.rot90(m, k=3) for item in sublist]
    
    tmp = f[:]
    f[2], f[5], f[8] = d[2], d[5], d[8]
    d[2], d[5], d[8] = b[6], b[3], b[0]
    b[6], b[3], b[0] = u[2], u[5], u[8]
    u[2], u[5], u[8] = tmp[2], tmp[5], tmp[8]
    
    result = [u, [l,f,r_new,b], d]
    return result
```
